```python
import libertas
import mesa
from mesa_llm.reasoning.cot import CoTReasoning
```

```python
model = mesa.Model()
```

```python
worker = libertas.agents.Worker(
    model,
    CoTReasoning,
    "ollama/llama2"
)
```

```python
pod = libertas.agents.Pod(
    model,
    CoTReasoning,
    "ollama/llama2"
)

pod.setup_pod(
    resources={"Money": 1000},
    workers=[worker],
    min_workers=1
)
```
